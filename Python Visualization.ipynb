# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Line Plot
x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

```python
#Scatter Plot
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)

plt.scatter(x,y,c=colors, cmap='viridis')
plt.colorbar(label = 'Color Values')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot')
plt.show()
```

```python
#Bar Plot
categories = ['A','B','C']
values = [10,25,30]

plt.bar(categories,values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Plot')
plt.show()
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,label='sin(x)')
plt.plot(x,y2,label='cos(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with legends')
plt.show()
```

```python
#Subplot
x=	np.linspace(0,10,100)
y1 = np.sin(x)
y1 = np.cos(x)

fig, (ax1,ax2) = plt.subplots(2,1)
ax1.plot(x,y1)
ax1.set_title('sin(x)')
ax2.plot(x,y2)
ax2.set_title('cos(x)')
plt.show()
```

# Plotly

```python
#Line Plots
import plotly.graph_objs as go

x	= np.linspace(0,10,100)
y = np.sin(x)
fig = go.Figure(data=go.Scatter(x=x,y=y, mode='lines', name='sin(x)'))
fig.update_layout(title='Simple Line Plot', xaxis_title='X-axis', yaxis_title='Y-axis')
fig.show()
```

```python
#Scatter Plot

x=	np.random.rand(100)
y=	np.random.rand(100)
fig = go.Figure(data=go.Scatter(x=x, y=y, mode='markers', name='Random Scatter Plot'))
fig.update_layout(title='Simple Scatter Plot', xaxis_title='X-axis', yaxis_title='Y-axis')
fig.show()
```

```python
#Bar Plot

categories = ['A','B','C']
values = [10,20,15]
fig = go.Figure(data=go.Bar(x=categories, y=values))
fig.update_layout(title='Simple Bar Plot', xaxis_title='Categories', yaxis_title='Values')
fig.show()
```

```python
x= np.linspace(0,10,100)
y = np.sin(x)
fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines', name='sin(x)'))
fig.update_layout(title='Line Plot of sin(x)',	xaxis_title='X-axis', yaxis_title='Y-axis')
fig.show()
```

```python
#Subplots
import plotly.subplots as sp

x= np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
fig = sp.make_subplots(rows=2, cols=1)
fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name='sin(x)'), row=1, col=1)
fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='cos(x)'), row=2, col=1)
fig.update_layout(title='Subplots of sin(x) and cos(x)')
fig.show()
```

# Practice Questions

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn	as sns
import plotly.graph_objs as go
```

```python
#Q1- Line Plot: Create a line plot of the functions y = x and y = x2 on the same graph. Add a legend to differentiate between the two lines.

x = np.linspace(0,10,100)
y1 = x
y2 = x**2

plt.figure()

plt.plot(x,y1,label='y=x')
plt.plot(x, y2, label='y=x^2')

plt.legend()

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#Q2- Scatter Plot: Create a scatter plot using random data for x and y. Color the points based on their value and add a color bar.

x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)

plt.scatter(x,y,c=colors, cmap='viridis')
plt.colorbar(label='Color Values')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#Q3- Subplots: Create a 2x2 grid of subplots. Each subplot should contain a different plot type (line plot, scatter plot, bar plot, and histogram) with some random data.

x = np.linspace(0,10,100)
y1 = np.random.rand(100)
y2 = np.sin(x)
categories = ['A','B','C']
values = [10,20,15]

fig, axs = plt.subplots(2,2)

axs[0,0].plot(x,y2)
axs[0,0].set_title('Line Plot')

axs[0,1].scatter(x,y1)
axs[0,1].set_title('Scatter Plot')

axs[1,0].bar(categories,values)
axs[1,0].set_title('Bar Plot')

axs[1,1].hist(y1, bins=10)
axs[1,1].set_title('Histogram')


plt.tight_layout()
plt.show()
```

```python
#Q4- Bar Plot: Create a bar plot showing the average monthly temperatures of a city. Add error bars to represent the standard deviation of the temperatures.

months = np.arange(1,13)
temperatures = np.random.randint(75,120,12)

plt.figure()
plt.bar(months, temperatures, yerr=np.std(temperatures))
plt.xlabel('Months')
plt.ylabel('Temperature')
plt.title('Average Monthly Temperatures')
plt.show()
```

```python
#Q5- Histogram: Create a histogram of 1000 data points drawn from a normal distribution. Add a KDE overlay to the histogram

data = np.random.randn(1000)

sns.histplot(data, bins=30, kde=True)
plt.show()
```
